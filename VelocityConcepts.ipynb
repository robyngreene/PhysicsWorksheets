## Scalars

A **scalar** quantity has only a magnitude (size). Examples of scalar quantities:
- mass
- distance
- speed
- density
- temperature

If we have two scalar quantities, they can be added or subtracted and it still makes sense:
1. plant A has a mass of 300g and plant B has a mass of 200g. The mass of the 2 plants together is simply <code>mass of plant A + mass of plant b</code>. In this case: $total mass = 300g + 200g = 500g$.
2. Ben walks 20m and stops for a snack. He then walks 100m. Ben has walked $20m + 100m = 120m$ in total.
3. Sonny walks 1 mile down the road, then walks back to his house. What distance has Sonny walked? (1 mile out and 1 mile back - 2 miles total.

Simple enough!

## Vectors

Vector quantities differ from scalars in that they have both magnitude <strong>and direction</strong>.

Examples:
- force - size in newtons, and direction
- displacement - distance and direction
- velocity - speed and direction
- acceleration - amount ($m/{s}^2$) and direction

# Contact and non-contact forces

In the box example, the force exerted involved physically touching the box. This is a contact force. Other examples of contact forces include:
- Tension (cable holding a ceiling lamp must involve some force upwards to counteract gravity on the lamp)
- Friction (force slowing an object down as it slides down a slope)

We wont focus too much on this right now, we'll come back to it when we focus on forces in more detail. If you are feeling keen, check out the [bitesize page on contact forces](https://www.bbc.co.uk/bitesize/guides/zyxv97h/revision/1)

<strong>Non-contact forces</strong> are more interesting for the current question!

These are forces between two objects that are not physically touching each other. Examples include magnetic force, electrostatic force and (drumroll please)...<strong>gravitational force</strong>.

# Gravitational force

Gravitational force is experienced by any mass in a <em>gravitational field</em>. The masses in question are attracted towards each other by <em>gravitational force</em>.

Any object with mass produces a gravitational field. The more mass, the greater the gravitational field will be.

# Newton's first law

An object travelling at a certain velocity will continue to travel at that velocity unless acted upon by a resultant force.

# Newton's second law

$$F = m a$$

Force can be calculated by multiplying mass by acceleration.

This is a bit of an aside, we'll revisit this later. It is important that you notice that there is a link between the force of gravity and acceleration.

Since gravity acts inwards towards the centre of the earth, the force of gravity causes objects with mass to accelerate towards the (centre of the) Earth.

Here are the steps to solve the question:

First step - the satellite is changing direction as it orbits:
- The satellite is orbiting Earth. 
- The satellite is in the Earth's gravitational field and therefore experiences force of gravity towards earth.
- The gravitational force pulls the satellite inwards, towards the Earth. 



- The force isn't strong enough to pull the satellite into the Earth, but is strong enough to change it's direction.
- The gravitational force will be the same (amount) as the satellite orbits.
- Therefore the satellite is always being pulled in towards the earth - it is always changing direction.

Second step - explain why this means a change in velocity:
- The speed of the satellite is always the same as it orbits.
- The satellite is constantly changing direction
- Velocity is a vector quantity - a change in direction (even if there is no change in speed) means a change in velocity.

Answer:

The force of gravity causes the satellite to accelerate towards the Earth. The acceleration causes a change in direction. Since velocity takes both speed and direction into account, the velocity of the satellite changes.
