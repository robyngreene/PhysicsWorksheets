### Gravity

- Objects with mass produce a gravitational field.
- More mass - greater gravitational field

- **Weight** - force acting on object due to gravity (N)
    - Acts towards centre of the field
    - weight is proportional to mass
    
Weight = mass * gravitational field strength

$$W = mg$$

Gravitational field strength (g) measured in newtons per kilo (N/kg):

For Earth: $g = 9.8 N/kg$<br>
For Moon: $g = 1.6 N/kg$
